```python
import os
import cv2
import numpy as np
from ultralytics import YOLO
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
imgs = [os.path.join('sub\data\Task5 data', path ) for path in os.listdir('sub\data\Task5 data')]
```

# Loading a custom YOLOV8-SEG Model

```python
model = YOLO("best.pt")
```

# Performing Segmentation on all the images

```python
for i in imgs:
    model(i,save = True)
    clear_output()
```

```python
predictions = [os.path.join('runs\segment\predict', path ) for path in os.listdir('runs\segment\predict')]
```

# Plotting a single image

```python
img = cv2.imread(predictions[6])
plt.imshow(img)
```

# Plotting all the Images

```python
plt.figure(figsize=(20,10))
columns = 10
for i, image in enumerate(predictions):
    plt.subplot(int(len(predictions) / columns + 1), columns, i + 1)
    plt.imshow(cv2.imread(image))
    plt.xticks()
    plt.yticks()
```
